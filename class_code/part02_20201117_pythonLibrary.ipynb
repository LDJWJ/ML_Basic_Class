```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,14,200) # 0부터 14까지 동일한 간격의 값 100개 만들기
x
```

```python
# x의 값을 이용하여 y값 (sin) 구하기
y1 = np.cos(x)
y2 = np.sin(x)

# 그래프 그리기
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
## 그래프에 범례 표시해보기
# x의 값을 이용하여 y값 (sin) 구하기
y1 = np.sin(x)
y2 = np.sin(x)*2
y3 = np.sin(x)*4

# import matplotlib.pyplot
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()

plt.title("sin plot")
```
